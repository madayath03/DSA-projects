# Distribution of sepal length

```python
plt.figure(figsize=(4,4))
plt.hist(data['SL'],color='green', rwidth=0.6)
plt.title('Distribution of Sepal length')
plt.xlabel('Sepal length')
plt.ylabel('Frequency')
```

# Distribution of sepal width

```python
plt.figure(figsize=(4,4))
plt.hist(data['SW'],color='red', rwidth=0.6)
plt.title('Distribution of Sepal width')
plt.xlabel('Sepal width')
plt.ylabel('Frequency')
```

# Distribution of petal length

```python
plt.figure(figsize=(4,4))
plt.hist(data['PL'],color='blue', rwidth=0.6)
plt.title('Distribution of petal length')
plt.xlabel('Petal length')
plt.ylabel('Frequency')
```

# Distribution of petal width

```python
plt.figure(figsize=(4,4))
plt.hist(data['PW'],color='yellow', rwidth=0.6)
plt.title('Distribution of petal width')
plt.xlabel('Petal width')
plt.ylabel('Frequency')
```
